Model 

```python
from sklearn import *
from sklearn.model_selection import train_test_split
```

```python
lnm = linear_model.LinearRegression()
```

```python
todrop=["MA_4","MA_10","MA_75","Close"]
```

```python
X=data.drop(todrop,1)
y=data["Close"]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.3,random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lnm.fit(X_train,y_train)
```

```python
lnm.predict(X)
```

```python
print("train accuracy :",lnm.score(X_train,y_train),"\ntest accuracy :",lnm.score(X_test,y_test))
```

```python
# Conclusion : Linear Regression is not an apt model since we see low test accuracy.
```

```python
plt.plot(y,X,"bo")
plt.plot(y,lnm.predict(X),"ro")
```

```python
rf=ensemble.RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
rf.fit(X_train,y_train)
```

```python
print ("train accuracy :",rf.score(X_train,y_train),"\ntest accuracy :",rf.score(X_test,y_test))
```

```python
# Conclusion : Random Forest is a better model that fits the data.
```

```python
rf.predict(X)
```

```python
plt.plot(y,X,"bo")
plt.plot(y,rf.predict(X),"ro")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(max_depth=10,random_state=0)
```

```python
dt.fit(X_train,y_train)
```

```python
print ("train accuracy :",dt.score(X_train,y_train),"\ntest accuracy :",dt.score(X_test,y_test))
```

```python
# Conclusion : Decision Tree is a better model than Linaer model bUt Random Forest is the best of these.
```
